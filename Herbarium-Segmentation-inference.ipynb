```python
import os
import matplotlib.pyplot as plt
from PIL import Image
base_dir = '/content/drive/.shortcut-targets-by-id/1nwtjFNI3EBmn_MJq6kZJINtl2OkZIgfW/Segmentation'

images_dir = os.path.join(base_dir, 'Images_test/')
predicted_dir = os.path.join(base_dir, 'Predicted_test_Black/')

# List the images in the folders.
image_files = sorted(os.listdir(images_dir))
predicted_files = sorted(os.listdir(predicted_dir))

# Assurez-vous que les noms correspondent et sont en ordre
image_files = [file for file in image_files if file.endswith('.jpg')]
predicted_files = [file for file in predicted_files if file.endswith('.jpg')]
```

```python
def display_images(images_dir, predicted_dir, image_files, num_images=3):
    plt.figure(figsize=(10, num_images * 5))

    for i, image_file in enumerate(image_files[:num_images]):
        # Ensure that the names match and are in order.
        img_path = os.path.join(images_dir, image_file)
        # Loading the original image
        original_image = Image.open(img_path)

        # Loading the predicted mask
        mask_path = os.path.join(predicted_dir, image_file)  # Ensure that the names match exactly
        predicted_image = Image.open(mask_path)

        # Displaying the original image
        plt.subplot(num_images, 2, 2*i + 1)
        plt.title('Original Image')
        plt.imshow(original_image)
        plt.axis('off')

        # Displaying the predicted mask
        plt.subplot(num_images, 2, 2*i + 2)
        plt.title('Predicted Mask')
        plt.imshow(predicted_image)
        plt.axis('off')

    plt.show()
```

```python
display_images(images_dir, predicted_dir, image_files, num_images=11)  # You can change 'num_images' to display more images.
```
